### Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

```python
# importing numpy and pandas
import numpy as np
import pandas as pd
```

```python
# creating array with random function
randnum=np.random.randint(1,501,500)
```

```python
# reshaping the array to [10,500]
my_array=randnum.reshape((10,50))
```

```python
my_array
```

```python
# Loading as dataframe
my_df=pd.DataFrame(my_array)
```

```python
my_df
```

### Calculate the sum of each row; the sum of each column; and the sum of all entries.

```python
# sum of each rows
my_df.sum(axis=1)
```

```python
# sum of each columns
my_df.sum(axis=0)
```

```python
# sum of all entries
sum(my_df.sum(axis=1))
```

### Determine the minimum; maximum; average; and standard deviation of all entries.

```python
# minimum row wise
my_df.min(axis=1)
```

```python
# minimum column wise
my_df.min(axis=0)
```

```python
# minimum of entire entries
my_df.to_numpy().min()
```

```python
# maximum row wise
my_df.max(axis=1)
```

```python
# maximum column wise
my_df.max(axis=0)
```

```python
# maximum of entire entries
my_df.to_numpy().max()
```

```python
# average row wise
my_df.mean(axis=1)
```

```python
# average column wise
my_df.mean(axis=0)
```

```python
# average of entire entries
my_df.to_numpy().mean()
```

```python
# sd row wise
my_df.std(axis=1)
```

```python
# sd column wise
my_df.std(axis=0)
```

```python
# sd of entire entries
my_df.to_numpy().std()
```
