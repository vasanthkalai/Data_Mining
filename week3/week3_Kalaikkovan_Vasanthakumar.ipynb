### A. Display the same analysis (of the same data) using 3 different charts (e.g.,  a bar chart, a line chart and a pie chart)
### B. Use appropriate, complete, professional labeling.

```python
# importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# gettig the data
df=pd.read_excel('Nifty_IT_Nov_2020.xlsx')
```

```python
df.head()
```

#### Line Chart

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(df['Date'],df['Close'])

ax.set(xlabel='Date', ylabel='Close Price',
       title='Nifty IT sector Nov 2020')
ax.grid()

plt.show()
```

#### Bar Chart

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.bar(df['Date'],df['Close'])

ax.set(xlabel='Date', ylabel='Close Price',title='Nifty IT sector Nov 2020')

plt.show()
```

#### Pie Chart

```python
import numpy as np

fig, ax = plt.subplots(figsize=(16,8))
def absolute_value(val):
    a  = np.round(val/100.*df['Close'].sum(), 0)
    return a

ax.pie(df['Close'], labels=df['Date'],autopct=absolute_value)
ax.set(title='Nifty IT sector Nov 2020')
plt.show()
```
